```python
from dusty import data as D
from dusty import util as U
```

load Nihao galaxies and SEDs

```python
hello = D.Nihao()
filters = U.ugrizJ()
```

```python
hello.prop
```

```python
fig = plt.figure(figsize=(10,10))
sub = fig.add_subplot(111)
for z in [0., 0.4, 0.8]: 
    sub.plot(hello.wave * (1+z), hello.seds[0] + z * 1e-4)
    _maggies = hello.maggies(z)
    sub.scatter(filters.effective_wavelengths, _maggies[0,:]/2.5e5 + z * 1e-4)
sub.set_xlabel('wavelength', fontsize=25)
sub.set_xlim(2e3, 1e4)
sub.set_ylim(0., 1e-4)
```

```python
fig = plt.figure(figsize=(10,10))
sub = fig.add_subplot(111)
for z in [0., 0.4, 0.8]: 
    sub.plot(hello.wave * (1+z), hello.seds[-10])
    _maggies = hello.maggies(z)
    sub.scatter(filters.effective_wavelengths, _maggies[-10,:]/2.5e5)# + z * 1e-4)
sub.set_xlabel('wavelength', fontsize=25)
sub.set_xlim(2e3, 1e4)
sub.set_ylim(0., 0.05)
```
